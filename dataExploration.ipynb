```python
# count proposals per year
# df['publishDate'].groupby([df.publishDate.dt.year]).agg('count')
print(df.dtypes)
df['publishDate'] = pd.to_datetime(df['publishDate'], format='%Y-%m-%d %H:%M:%S')
```

```python
# how does the participation look over the years

df['publishDate'].groupby([df.publishDate.dt.year]).agg('count')
```

```python
# plot 
fig, axs = plt.subplots(figsize=(12, 4))
df['publishDate'].groupby([df.publishDate.dt.year]).agg('count').plot(
    kind='bar', rot=0, ax=axs
)
plt.xlabel("Year of Publication");  # custom x label using Matplotlib
plt.ylabel("Number of Proposals");
```

```python
# are there any seasonal trends?
df['publishDate'].groupby([df.publishDate.dt.year, df.publishDate.dt.month]).agg('count')[:12]
```

```python
# are there seasonal trends? observe the first 36 months
fig, axs = plt.subplots(figsize=(40, 12))
df['publishDate'].groupby([df.publishDate.dt.year, df.publishDate.dt.month]).agg('count')[:36].plot(
    kind='bar', rot=0, ax=axs
)
plt.xlabel("Month of Publication");  # custom x label using Matplotlib
plt.ylabel("Number of Proposals")

# was passierte 2017, Oktober, November? Woher der Sprung?
# Gibt es einen saesonalen Trend?
```

```python
# Who is the top poster for prosals? Is there someone?
# fig, axs = plt.subplots(figsize=(20, 12))
df['proposer'].groupby([df.proposer]).agg('count').sort_values(ascending=False)[:20]
```

```python
# are there some people good at conveying the message
# let's look at the accumulated votes of top 20
#df.columns
df[['proposer', '#Votes']].groupby([df.proposer]).agg({'proposer':'count', '#Votes': 'sum'}).sort_values(by=["#Votes"],ascending=False)[:20]
# We can say, frequency does not match popularity
# Maybe some people will just use the platform to vent
```

```python
# How many proposals did get through
# How many did just missed their threshold by little
# df[["#Votes", "MinVotesNecessary"]]
#(df['#Votes']<=df['MinVotesNecessary']).agg('count')
#df.where(df["#Votes"]>=df["MinVotesNecessary"])
df["DeltaVotes"]=df["#Votes"]-df["MinVotesNecessary"]
```

```python
# How many votes did make it through?

# Where do we have more votes than threshold
len(df[df["DeltaVotes"]>=0])
```

```python
# we also need to consider that votes can't be zero, b
successfulProposals = df[(df["DeltaVotes"]>=0) & (df["MinVotesNecessary"]!=0)]
len(successfulProposals)

# there were 275 successful proposals
```

```python
# What's the percentage of success?
len(successfulProposals)/len(df)*100
```
